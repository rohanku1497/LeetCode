```python
nums=[-1,0,1,2,-1,-4]
out=[]
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        for k in range(j+1,len(nums)):
            if i!=j and i!=k and j!=k:
                if nums[i] + nums[j] + nums[k] == 0:
                    out.append([nums[i],nums[j],nums[k]])

out=[sorted(o) for o in out]  
outn=[]
for o in out:
    if o not in outn:
        outn.append(o)

print(outn)                    
```
