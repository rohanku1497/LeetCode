Question

Write a program that accepts a sequence of whitespace-separated words as input, removes duplicate words, sorts the remaining words alphanumerically, and prints them as a whitespace-separated sequence.

Input

A sequence of whitespace-separated words.

Output

The unique words sorted alphanumerically and printed as a whitespace-separated sequence.

Example

Input

hello world and practice makes perfect and hello world again

Output

again and hello makes perfect practice world

```python
s="hello world and practice makes perfect and hello world again"
x=s.split(" ")
out=[]
for i in x:
    if i not in out:
        out.append(i)
out.sort()  
y=' '.join(out)   
y   
```
