Write a program that accepts a sentence as input, counts the frequency of each word, and prints each word with its frequency after sorting the words alphanumerically.

Input

A sentence containing words separated by spaces.

Output

Each unique word followed by its frequency in the format:

<word>:<count>

Print the words in alphanumerical order.

```python
s="New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3"
s=s.split(" ")
s.sort()
s
```

```python
s="New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3"
s=s.split(" ")
s.sort()
d={}
for i in s:
    if i in d.keys():
        d[i]+=1
    else:
        d[i]=1    
```

```python
print(d)
```
