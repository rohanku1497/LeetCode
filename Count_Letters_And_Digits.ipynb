Question

Write a program that accepts a sentence as input and calculates the number of letters and digits in the sentence.

Input

A sentence.

Output

Print the number of letters and digits in the following format:

LETTERS <count>
DIGITS <count>

Example

Input

hello world! 123

Output

LETTERS 10
DIGITS 3

```python
s='hello world! 123'
s
```

```python
import re
c=0
n=0
for i in s:
    if re.match(r'[A-Za-z]',i):
        c=c+1
    elif  re.match(r'[0-9]',i): 
        n=n+1
print(c,n)        
```
