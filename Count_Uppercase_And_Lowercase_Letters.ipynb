Question

Write a program that accepts a sentence as input and calculates the number of uppercase and lowercase letters in the sentence.

Input

A sentence.

Output

Print the number of uppercase and lowercase letters in the following format:

UPPER CASE <count>
LOWER CASE <count>

Example

Input

Hello world!

Output

UPPER CASE 1
LOWER CASE 9

```python
s="Hello world!"
s
```

```python
u=0
l=0
for i in s:
    if i.isupper():
        u=u+1
    elif i.islower():
        l=l+1
print(u,l)
```
