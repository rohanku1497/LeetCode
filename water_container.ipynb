```python
h = [1,8,6,2,5,4,8,3,7]
out=[]
for i in range(1,len(h)):
    for j in range(i+1,len(h)):
        if h[i]<h[j]:
            height=h[i]
        elif h[j]<h[i]:
            height =h[j]
        else:
            height=h[i]
        area= height*(j-i)
        out.append(area)
out.sort()        
print(out[-1])
```
