```python
net=0
while True:
    transaction=input('Enter transaction type')
    if net==0 and (transaction == 'W' or transaction=='w'):
        print('No balance')
        break
    elif transaction == 'D' or transaction=='d':
        amount=int(input('Enter amount'))
        net=net+amount
    elif transaction == 'w' or transaction=='W':
        amount=int(input('Enter amount'))
        net=net-amount
    else:
        break    
print(net)    



```
