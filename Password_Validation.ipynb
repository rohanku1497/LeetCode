Question

Write a program that accepts a comma-separated sequence of passwords and prints only the passwords that satisfy the following conditions:

At least one lowercase letter (a-z)
At least one uppercase letter (A-Z)
At least one digit (0-9)
At least one special character from ($, #, @)
Minimum length: 6
Maximum length: 12
Input

A comma-separated sequence of passwords.

Output

The valid passwords printed as a comma-separated sequence.

Example

Input

ABd1234@1,a F1#,2w3E*,2We3345

Output

ABd1234@1

```python
import re
while True:
    pwd = input('enter password')
    if re.search(r'[A-z]',pwd) and re.search(r'[a-z]',pwd) and re.search(r'[0-9]',pwd)and re.search(r'[$#@]',pwd):
        print(pwd)
        break
    else:
        print('Wrong pwd')
        break


pwd
```

```python
def putNumbers(n):
    i = 0
    while i<n:
        j=i
        i=i+1
        if j%7==0:
            yield j
```
