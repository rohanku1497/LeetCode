```python
intervals = [[1,3],[2,6],[8,10],[15,18]]
```

```python
merged=[]
for i in intervals:
    if not merged or i[0]>merged[-1][1]:
        merged.append(i)
    else:
        merged[-1][1]=max(merged[-1][1],i[1])    
print(merged)
```
